# Final Load Preparation

This notebook prepares the final dataset for Tableau visualization and creates optimized data exports for dashboard creation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

print('Final Load Preparation Started...')
print(f'Timestamp: {datetime.now()}')
```

```python
# Load the master dataset
df = pd.read_csv('../data/processed/master_dataset.csv')

print(f'Original Dataset Shape: {df.shape}')
print(f'Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB')
print(f'Columns: {len(df.columns)}')
```

## Data Quality Assessment

```python
# Final data quality check
print('FINAL DATA QUALITY ASSESSMENT:')
print('=' * 50)

# Missing values
missing_summary = df.isnull().sum()
missing_pct = (missing_summary / len(df)) * 100

print('\nMISSING VALUES SUMMARY:')
for col in missing_summary[missing_summary > 0].index:
    print(f'  {col}: {missing_summary[col]:,} ({missing_pct[col]:.1f}%)')

# Data types
print(f'\nDATA TYPES:')
print(df.dtypes.value_counts())

# Duplicates
duplicates = df.duplicated().sum()
print(f'\nDUPLICATE ROWS: {duplicates:,}')
```

## Export Optimized Dataset

```python
# Select key columns for main dashboard
tableau_columns = [
    'order_id', 'order_status', 'order_purchase_timestamp',
    'customer_id', 'customer_unique_id', 'customer_state', 'customer_city',
    'product_id', 'product_category_name', 'product_category_name_english',
    'price', 'payment_type', 'payment_value', 'payment_installments',
    'review_score', 'sentiment', 'is_negative_review',
    'delivery_time_days', 'is_delivered_clean', 'freight_value', 'item_total_value'
]

# Filter columns that exist in the dataset
tableau_columns = [col for col in tableau_columns if col in df.columns]

# Create optimized dataset
df_tableau = df[tableau_columns].copy()

print(f'Optimized dataset shape: {df_tableau.shape}')
print(f'Columns: {len(df_tableau.columns)}')
print(f'Memory usage: {df_tableau.memory_usage(deep=True).sum() / 1024**2:.2f} MB')
```

```python
import os

# Create output directory
if not os.path.exists('../tableau'):
    os.makedirs('../tableau')
    print('Created tableau directory')

# Export main Tableau dataset
df_tableau.to_csv('../tableau/main_dashboard_data.csv', index=False)
print(f'Exported main dashboard data: {df_tableau.shape}')

print('\nFinal Load Preparation Completed Successfully!')
```
